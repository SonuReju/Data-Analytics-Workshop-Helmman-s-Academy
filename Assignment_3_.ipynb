**Reading Score vs. Writing Score**

```
xData = [x for x in range(10)]
yData = [y for y in range(20)]
sns.lineplot(y="writing score" , x="reading score" , data=Studentperf)
```

**Math Score vs. Writing Score**

```
Studentperf.plot.scatter("math score","writing score", figsize = (9,3))
```

```
xData = [x for x in range(10)]
yData = [y for y in range(20)]
sns.lineplot(y="writing score",x="math score" , data=Studentperf)
```

**Math Score vs. Reading score**

```
Studentperf.plot.scatter("math score","reading score", figsize = (9,3))
```

```
xData = [x for x in range(10)]
yData = [y for y in range(20)]
sns.lineplot(x= "math score" , y="reading score", data=Studentperf)
```

```
from matplotlib import pyplot
size=(15,3)
ax=pyplot.subplots(figsize=size)
sns.countplot(x="parental level of education", data=Studentperf)
```

**Conclusions**

We can see that reading score and writing score are directly proportional, as expected. The same is true for the math vs. reading score graph and math vs. writing score graph.
In the parental level of education graph, we can see that the parents of most students only have some college education and and the children of master's degree holders are the lowest number of students.
